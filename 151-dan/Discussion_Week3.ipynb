# Univariate Gaussian

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

h = [186, 176, 158, 180, 186, 168, 168, 164, 178, 170, 189, 195, 172,
     187, 180, 186, 185, 168, 179, 178, 183, 179, 170, 175, 186, 159,
     161, 178, 175, 185, 175, 162, 173, 172, 177, 175, 172, 177, 180]

h.sort()
mean = np.mean(h)
std = np.std(h)
print ("Mean = ", mean, ", Std = ", std)
pdf = stats.norm.pdf(h, mean, std)
plt.plot(h, pdf)
plt.show()
```

```python
h1 = [x+20 for x in h]
h1.sort()
mean1 = np.mean(h1)
std1 = np.std(h1)
print ("Mean1 = ", mean1, ", Std1 = ", std1)
pdf1 = stats.norm.pdf(h1, mean1, std1)
plt.plot(h1, pdf1, color="orange")
plt.show()
```

```python
plt.plot(h , pdf , label="class 0")
plt.plot(h1, pdf1, label="class 1")
plt.legend(loc = 1)
plt.xlabel('Feature Points')
plt.ylabel('PDF')
plt.show()
```

## Test Point Classification using Univariate Gaussians

```python
test = [182, 198, 159, 170, 196, 178, 181, 200, 0, 250]

test_pdf = []
test_pdf = [[x, stats.norm.pdf(x, mean, std), stats.norm.pdf(x, mean1, std1)] for x in test]
test_pdf = np.array(test_pdf, dtype=np.float32)
print (test_pdf)
#print (test_pdf, "\n")

preds = np.argmax(test_pdf[:,1:], axis = 1) 
for a,b in zip(test, preds):
    print (a,b)
```

# Bivariate Gaussian

```python
means = [0, 0]  
covs = [[4 , 0], 
        [0,  4]] 
```

## Computing PDF for Bivariate Gaussians

```python
x = [1,1]
diff = np.array(x)-np.array(means)
pdf = 0
print ("pdf = ", pdf)
Nr = np.exp(-0.5*np.matmul(np.matmul(diff.T, np.linalg.inv(covs)) , diff))
Dr = float(2*np.pi*np.sqrt(np.linalg.det(covs)))
pdf = Nr/Dr
print("pdf = ", pdf)
```

```python
from scipy.stats import multivariate_normal

pdf = multivariate_normal.pdf(x, means, covs);
print (pdf)
```
